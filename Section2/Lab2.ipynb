```python
import cv2
img=cv2.imread("windows.jpg",0)  # 0--> gray scale, 1 ---> read as colors, -1 -->more accurate
cv2.imshow('image',img)
cv2.waitKey(0)   # 0 to be shown untill press on any button on the keyboard and maybe time of millisecond of showing
cv2.destroyAllWindows()
```

```python
import cv2
img=cv2.imread("windows.jpg",-1)
cv2.imshow("image",img)
k=cv2.waitKey(0)
if k==27:
    cv2.destroyAllWindows()
elif k==ord('s'):
    cv2.imwrite("newWindows.png",img)
    cv2.destroyAllWindows()
```
